3.2  Prompt the user to enter an integer value for the month  (1..12) and read it in

3.3  Prompt the user to enter an integer value for the day  (1..31) and read it in

    y0 = y - (14 - m) // 12

    x = y0 + y0 // 4 - y0 // 100 + y0 // 400

```python
# Your code for exercise 3 goes here
# I wrote the first prompt for you to get you started

year = int(input('What is the year: '))
month = int(input('Enter the number of the month (1-12): '))
day = int(input('Enter the number day (1-31): '))

y0 = year - (14 - month) // 12

x = y0 + y0 // 4 - y0 // 100 + y0 // 400

m0 = month + 12 * ((14 - month) // 12) - 2

d0 = (day + x + (31 * m0) // 12) % 7

print(f'The day of the week is {d0}.')
#told to print out only as integer for assignment
```
