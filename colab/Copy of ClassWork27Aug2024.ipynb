```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

majors =['Biology', 'Data Sci', 'Psych', 'BioInf', 'Info Sys']
pos = np.arange(len(majors))
popularity = [4, 10, 5, 7, 1]

plt.bar(pos, popularity, align='center', color =['magenta', 'red', 'orange', 'yellow', 'green', 'blue', ] )
plt.xticks(pos, majors)
plt.ylabel('Major')
plt.title('Majors in our Math 215 Class', alpha=0.8)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

buckets =['blue', 'red', 'green', 'light blue', 'purple']
pos = np.arange(len(buckets))
population = [4, 6, 5, 3, 2]

plt.bar(pos, population, align='center', color =['blue', 'red', 'green', 'cyan', 'violet'])
plt.xticks(pos, buckets)
plt.ylabel('Favorite Color')
plt.title('Favorite Colors in Math 215', alpha=0.8)

plt.show()
```
