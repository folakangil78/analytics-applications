Factor drill solution.  How should we handle prime numbers, and user entries of 1, or a == b?
Modify the code

```python
import random

num_problems = 5
print(f'Welcome to our factor drill program.  We will have {num_problems} exercises')
score = 0
for k in range(num_problems):
  c = random.randint(4, 100)
  print(f'Enter values for factoring a x b = {c}')
  a = int(input('a: '))
  b = int(input('b: '))
  round_score = 0
  if c % a == 0:
    print(f'{a} is a factor of {c}')
    round_score += 1
  else:
    print(f'{a} is not a factor of {c}')
  if a != b or a * b == c:
    if c % b == 0:
      round_score += 1
    if a * b == c:
      round_score += 2
  print(f'You earned {round_score} points this round')
  score += round_score
print(f'Your final score is {score} points')
```

Using random numbers:  (Like exercise for Sept 5 but using the first quadrant only)  Randomly generate the coordinates for a point in x-y plane, with the restriction that x and y are both in the interval [0, 1).  (So, your program will generate two floating point numbers in the interval [0, 1)]).  Then test to see if the point falls inside or on the boundary of the unit circle.   Use this to compute a Monte Carlo approximation for the value of pi.  Here's the algorithm:

1.1.1  Throw n darts at the rectangle

1.1.2  Approximate pi by observing the (number of darts inside)/n  is approximately the same as (area of one quarter of the unit circle), so 4*((number of darts inside)/n  is approximately pi

1.2.  Approximate pi using a series:

pi/4 = 1 -1/3 + 1/5 - 1/7 + 1/9 - 1/11 + 1/13

Use a for loop to add the terms in the series.  How many terms in the series do you need to sum to get an approximation that is accurate to 4 decimal places?





```python
import random

inside_counter = 0
num_trials = 100

for k in range(num_trials):
  x = random.random()
  y = random.random()
  if ((x * x) + (y * y)) <= 1:
    inside_counter += 1

print(f'{4 * (inside_counter / num_trials)}')

```

```python
import random

num_trials = 1000
counter = 0
for k in range(num_trials):
  dice1 = random.randint(1,7)
  dice2 = random.randint(1,7)
  if dice1 + dice2 == 7:
    counter+=1
print(f'Probability of rolling a 7: {counter / num_trials}')
```
