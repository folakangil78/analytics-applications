Use a for-loop to sum a series

```python
# Geometric series with r = 1/2 converges fast to 2
num_terms = 200
running_sum = 0
for n in range(num_terms):
  next_term = 1/2**n
  running_sum += next_term

print(f'The sum is {running_sum} using {num_terms} terms')
```

```python
# Alternating harmonic series, converges slowly to ln(2)
num_terms = 200000
running_sum = 0
for n in range(1, num_terms):
  next_term = (-1)**(n-1) / n
  running_sum += next_term

print(f'The sum is {running_sum} using {num_terms} terms')
```

Use a for loop to repeat this code with increasing values of num_terms

```python
for num_terms in [20, 200, 2000, 20000, 200000]:
  running_sum = 0
  for n in range(1, num_terms):
    next_term = (-1)**(n-1) / n
    running_sum += next_term

  print(f'The sum is {running_sum} using {num_terms} terms')
```

A series that diverges

```python
# Harmonic series, diverges slowly
num_terms = 200000
running_sum = 0
for n in range(1, num_terms):
  next_term = 1 / n
  running_sum += next_term

print(f'The sum is {running_sum:.4f} using {num_terms:,} terms')

```

```python
for num_terms in [20, 200, 2000, 20000, 200000, 2000000]:
  running_sum = 0
  for n in range(1, num_terms):
    next_term = 1 / n
    running_sum += next_term

  print(f'The sum is {running_sum:.4f} using {num_terms:,} terms')
```

Dice problems

```python
import matplotlib.pyplot as plt
import numpy as np
import random

plt.figure()
sides = 10
buckets =[k for k in range(2, 2*sides + 1)]
pos = np.arange(len(buckets))
results = {k:0 for k in range(2, 2*sides + 1)}
num_trials = 100000
for k in range(num_trials):
  d1 = random.randint(1, sides)
  d2 = random.randint(1, sides)
  results[d1+d2] += 1
population = [results[k] for k in range(2, 2*sides + 1)]
plt.bar(pos, population, align='center', color =['blue', 'violet', 'red', 'orange',  'green', 'cyan' ] )
plt.xticks(pos, buckets)
plt.ylabel('Frequency')
plt.title(f'Distribution of dice rolls with {sides}-sided dice,  {num_trials:,} rolls', alpha=0.8)
print(f'Probability of rolling an eleven:  {results[11]/num_trials:.4f}')
plt.show()
```

Iterating over a string

```python
my_str = 'a cat and a dog'
print(f'len(my_str) = {len(my_str)}')
print(f'my_str.count("a") = {my_str.count("a")}')
for k in range(len(my_str)):
  print(f'k = {k}, my_str[{k}] = {my_str[k]}', end = ' $$ ')

print(f'my_str.find("a") = {my_str.find("a")}')
print(f'my_str.find("z") = {my_str.find("z")}')
answer = ''
for symbol in my_str:
  if symbol in 'abc':
    answer += symbol

print(f'answer = {answer}')
```
