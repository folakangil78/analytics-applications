Use a for-loop to sum a series

```python
# Example from last week's gw
for num_terms in [5, 50, 500, 1000]:
  running_sum = 0
  for n in range(num_terms):
    next_term = n/3**n
    running_sum += next_term

  print(f'The sum is {running_sum:.8f} using {num_terms:,} terms')
```

Student solution using sum and list comprehension

```python
answer = sum([n/(3**n) for n in range(1, 50)])
print(f'The sum is {answer:.8f} using {50} terms')
```

Iterating over a string, creating a new string during the iteration

```python
char_set = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
vowels = 'AEIOU'
consonants = ''
for letter in char_set:
  if letter not in vowels:
    consonants += letter
print (f'There are {len(consonants)} consonants:  {consonants}')
print (f'There are {len(vowels)} vowels:  {vowels}')
```

Write a hi-low guessing game using a while loop

1.  Generate a random integer n in the interval 0 <= n <= 100

2.  Repeatedly prompt user to guess the number until they get it correct

3.  For each wrong guess, give accurate feedback of "too high"  or "too low"

4.  Terminate only when the user gets the correct number.  Report how many guesses it took.

```python
#  Write this one together as a class
import random
answer = random.randint(0,100)
print(answer)

user_input = -1 #intentionally wrong
guess_count = 0
while(user_input != answer):
  user_input = int(input('Guess my number here: '))
  guess_count += 1
  if user_input < answer:
    print('Your guess is too low. Try again')
  elif user_input > answer:
    print('Your guess is too high. Try again')
  else:
    print(f'Correct answer in {guess_count} guesses')

print('Goodbye')

```
