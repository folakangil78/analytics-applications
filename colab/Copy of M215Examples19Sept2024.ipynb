Test if an integer is prime by looking for a factor (how can we improve this)

```python
# Correct but inefficient
n = 43
f = 2
while f < n and n % f !=0:
  f = f+1
print(f'n:  {n}, f: {f}')
if n == f:
  print(f'{n} is prime')
```

Improved prime test

2 is the only even prime.  

If n has a factor then it must have a factor <= sqrt(n)

```python
n = int(input('Enter a positive integer: '))
if n > 2 and n % 2 == 0:
  is_prime = False
  factor = 2
else:
  is_prime = True
  factor = 3
  while is_prime and factor * factor <= n:
    if n % factor == 0:
      is_prime = False
    else:
      factor += 2

if is_prime:
  print(f'{n:,} is prime')
else:
  print(f'{n:,} is not prime.  {factor:,} is a factor of {n:,}')
```

Caesar Cipher Solution with intermediate steps and variables

```python
# Caesar Cipher program for Math 215 Sept 19, 2024

char_set = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# Prompt user and read in an integer shift value, n.
n = 347

# Prompt user and read in the text to be encoded, store
# this in a variable called plaintext

plaintext = 'CaT or ZEBRA'  # or read it from user

plaintext = plaintext.upper()  # Make sure it is upper case

# Use indexing into charSet to convert a letter to a number
# Encode the plaintext using a caesar cipher with right shift n, creating ciphertext
ciphertext = ''
for symbol in plaintext:
  symbol_num = char_set.find(symbol)
  if symbol_num != -1:
    code_num = (symbol_num + n) % len(char_set)
    code_symbol = char_set[code_num]
    ciphertext += code_symbol
  else:
    ciphertext += symbol
  #print(f'symbol: {symbol}, symbol_num: {symbol_num}, ciphertext: {ciphertext} ')

print(f'ciphertext: {ciphertext}')

decode_text = ''
for symbol in ciphertext:
  symbol_num = char_set.find(symbol)
  if symbol_num != -1:
    decode_num = (symbol_num - n) % len(char_set)
    decode_symbol = char_set[decode_num]
    decode_text += decode_symbol
  else:
    decode_text += symbol

print(f'decode_text: {decode_text}')

if decode_text == plaintext:
  print('Success')
else:
  print('Fail')

```

Shorter solution

```python
# Caesar Cipher program for Math 215 Sept 19, 2024

char_set = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# Prompt user and read in an integer shift value, n.
n = 10

# Prompt user and read in the text to be encoded, store
# this in a variable called plaintext

plaintext = 'CaT or ZEBRA'  # or read it from user

plaintext = plaintext.upper()  # Make sure it is upper case

# Use indexing into charSet to convert a letter to a number
# Encode the plaintext using a caesar cipher with right shift n, creating ciphertext
ciphertext = ''
for symbol in plaintext:
  if symbol in char_set:
    ciphertext += char_set[(char_set.find(symbol) + n) % 26]
  else:
    ciphertext += symbol

print(f'ciphertext: {ciphertext}')

decode_text = ''
for symbol in ciphertext:
  if symbol in char_set:
    decode_text += char_set[(char_set.find(symbol) - n) % 26]
  else:
    decode_text += symbol

print(f'decode_text: {decode_text}')

if decode_text == plaintext:
  print('Success')
else:
  print('Fail')

```

Nested while loops example

```python
n = 5

row = 0
while row < n:
  print(f'{row}:', end = ' ')
  col = 0
  while col < row:
    print(f'{col} ', end = ' ')
    col += 1
  print()
  row += 1

```

1.  Partner work: Compute the reverse complement of a DNA string

```python
dna_str = 'ACTTGGTAAGTC'

list_dna = list(dna_str)

for i in range(len(list_dna)):
  if list_dna[i] == 'A':
    list_dna[i] = 'T'
  elif list_dna[i] == 'C':
    list_dna[i] = 'G'
  elif list_dna[i] == 'T':
    list_dna[i] = 'A'
  elif list_dna[i] == 'G':
    list_dna[i] = 'C'
print(list_dna)
```
