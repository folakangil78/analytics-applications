Converting seconds to Days + Hours + Minutes + Seconds

For example, 493,383 seconds = 5 days, 17 hours, 4 minutes and 23 seconds

```python
total_seconds = int(input("Enter an integer number of seconds"))
# To do:  complete this

days = total_seconds // (24*60*60)
seconds_left = total_seconds % (24*60*60)
hours = seconds_left // (60*60)
seconds_left = seconds_left % (60*60)
minutes = seconds_left // 60
seconds_left = seconds_left % 60

print(f'{total_seconds} = {days} days and {hours} hours and {minutes} minutes and {seconds_left} seconds')
```

```python
a = int(input('Enter a value for a: '))
b = int(input('Enter a value for b: '))
c = int(input('Enter a value for c: '))

discriminant = (b * b) - (4 * a * c)
if discriminant > 0:
  sqrt_root = math.sqrt((b * b) - (4 * a * c))
  x_top_fraction = -(b) - sqrt_root
  y_top_fraction = -(b) + sqrt_root
  bottom_fraction = 2 * a
  x_result = x_top_fraction / bottom_fraction
  y_result = y_top_fraction / bottom_fraction
else:
  print('Your discriminant was negative! You are unable to find the roots to ' +
        'the equation.')

print(f'X value of root: {x_result} and Y value of root: {y_result}')
```
