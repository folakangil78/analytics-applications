Frequency Table for dice rolls

```python
import random
roll_freq = {}
num_rolls = 100000
for k in range(num_rolls):
  result = random.randint(1,6) + random.randint(1, 6)
  if result in roll_freq:
    roll_freq[result] += 1
  else:
    roll_freq[result] = 1

print(roll_freq)
for r in roll_freq:
  print(f'{r} occurred {roll_freq[r]} times.')

print()

# for r in sorted(roll_freq):
#   print(f'{r} occurred {roll_freq[r]} times.')

# print(sorted(roll_freq.keys()))

# print(roll_freq.keys())
# print(roll_freq.values())
# print(sorted(roll_freq))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
sides = 6
rolls =list(sorted(roll_freq.keys()))
pos = np.arange(2, 2 + len(rolls))

#popularity = [roll_freq[r] for r in roll_freq]
popularity = [roll_freq[r] for r in rolls]

plt.bar(rolls, popularity, align='center', color =['brown', 'orange', 'yellow', 'green', 'blue'] )
plt.xticks(pos, rolls)
plt.ylabel('Frequency')
plt.title(f'Frequency of rolls of a pair of {sides}-sided dice, {num_rolls} trials', alpha=0.8)

plt.show()
```

Dictionary Example - organizing data

keys represent different traits of one user

```python
user_1 = {'first_name': 'Alice', 'last_name':'Smith', 'acct_balance': 525.00}
user_2 = {'first_name': 'Jose', 'last_name':'Ramirez', 'acct_balance': 450.00}

user_list = [user_1, user_2]
print(user_list)
# create 2 more users
# add them to the list
for k in range(0):
  fn = input('first name: ')
  ln = input('last name: ')
  ab = float(input('account balance: '))
  u = {'first_name':fn, 'last_name': ln, 'acct_balance': ab}
  user_list.append(u)
print(user_list)

# same thing, all in one line
for k in range(4):
  user_list.append({'first_name':input('first name: '), 'last_name': input('last name: '), 'acct_balance': float(input('account balance: '))})

print(user_list)
# print all the user last names from list
for u in user_list:
  print(u['last_name'])

# find user with the highest acct balance from list

current_max = 0
max_user = None

for u in user_list:
  if u['acct_balance'] > current_max:
    current_max = u['acct_balance']
    max_user = u

print(max_user)
```
