Letter frequency example from Tuesday, modified to read from a file

```python
letter_freq = {}

filename = 'NobelPrizeAI.txt'
#filename = 'Saccharomyces18S.txt'
with open(filename, 'r') as f:
  text_passage = f.read()

print(len(text_passage))

text_passage = text_passage.lower()
#print(text_passage)
alphabet = 'abcdefghijklmnopqrstuvwxyz'

letter_freq = {letter : 0 for letter in alphabet}
print(len(letter_freq))
# Start here
for symbol in text_passage:
  if symbol in alphabet:
    letter_freq[symbol] += 1

# Display a table of letter frequencies:
#print(letter_freq)

print(f'letter  frequency')
for letter in letter_freq:
  print(f'{letter:7}{letter_freq[letter]:4}')

# Find the largest and smallest keys
print(f'max(letter_freq) is {max(letter_freq)}')
print(f'min(letter_freq) is {min(letter_freq)}')

# Find the keys with the largest and smallest values
print(f'max(letter_freq, , key = letter_freq.get) is {max(letter_freq, key = letter_freq.get)}')
print(f'min(letter_freq, key = letter_freq.get) is {min(letter_freq, key = letter_freq.get)}')
```

Bar Graph of letter frequencies

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

popularity = [letter_freq[sym] for sym in alphabet]
#pos = np.arange(len(alphabet))
#plt.bar(pos, popularity, align='center', color =[ 'green', 'blue'] )
plt.bar(list(alphabet), popularity, align='center', color =[ 'green', 'blue'] )
#plt.xticks(pos, list(alphabet))
plt.ylabel('Frequency')
plt.xlabel('Alphabet')
plt.title(f'Frequency of letters in {filename}', alpha=0.8)

plt.show()
```

Create a list of student first names and write it to a file

Step 1:  Remove the header

Step 2:  Remove the double quotes, tabs and endlines

Step 3:  Locate first names

Step 4:  Write to a file

```python
filename = 'M215Roster.txt'
with open(filename, 'r') as f:
  class_list = f.readlines()

class_list = class_list[1:] # slice from 1 to end - gets rid of header
class_list = [item.strip('\t\n"') for item in class_list]
print(class_list)

first_names = [entry[entry.find(',') + 2:] for entry in class_list]
print(first_names)

outfilename = 'M215FirstNames.txt'
with open(outfilename, 'w') as outf:
  for name in first_names:
    outf.write(name + '\n')

print(f'Done creating {outfilename}')

outfilename2 = 'M215FirstNames2.txt'
with open(outfilename2, 'w') as outf:
  outf.writelines(first_names)

print(f'Done creating {outfilename2}')
```

```python
import csv
filename = 'CTA-Ridership.csv'
CTArides2024 = []
with open(filename, 'r') as csvfile:
    my_reader = csv.reader(csvfile, delimiter=',')
    row_num = 1
    for row in my_reader:
      if row_num > 1 and row[0][-4:] == '2024':
          CTArides2024.append(row)

      if row_num > 1 and row[0][-4:] != '2024' :
        break
      row_num += 1
print(len(CTArides2024))
```
