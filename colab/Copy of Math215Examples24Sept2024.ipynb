Dynamic typing and polymorphism

```python
# Demonstrate parameter passing
import math

def f(x):  # what is the type of x? x only exists inside f
    print('Inside function f, type(x) = ', type(x))
    return 3*x

def main():
    print('f(5) = ', f(5))
    print('f(math.pi) = ', f(math.pi))
    print('f(\'abc\') = ', f('abc'))
    print('f([5, 6, 7]) = ', f([5, 6, 7]))
    print('f((5, 6, 7)) = ', f((5, 6, 7)))
    print('f(True) = ', f(True))
    print('f(False) = ', f(False))
    print('int(True) = ', int(True))
    print('int(False) = ', int(False))
    #print('x = ', x)


main()
```

Function calls with mutable types

```python
def replace_first_v1(aseq, newVal):  # input parameters : aseq, newVal
    # This works for lists, fails for strings
    # Notice there is no return value
    aseq[0]=newVal  # assignment is illegal for strings - they are immutable

def replace_first_v2(aseq, newVal):  #inpute parameters aseq, newVal
    aseq = newVal + aseq[1:]  # Creating a new string
    return aseq

def main():
    myList = [1,2, 3]
    print(f'Initially myList is : {myList}')
    replace_first_v1(myList, 'A')
    print(f' after the call to replace_first_v1 myList is : {myList}')
    print()

    myStr = 'xyz'
    print(f'Initially myStr is : {myStr}')
    answer = replace_first_v2(myStr, 'A')
    print(f' after the call to replace_first_v2 myStr is : {myStr}')
    print(f' after the call to replace_first_v2 answer is : {answer}')
    print()

    myList = [1,2, 3]
    print(f'Initially myList is : {myList}')
    ans = replace_first_v2(myList, ['A'])
    print(f' after the call to replace_first_v2 myList is : {myList}')
    print(f' after the call to replace_first_v2 ans is : {ans}')
    print()

main()
```

Functions as parameters to other functions

```python
import math
def sum_series(start, stop, a):
  # Add the terms a(start) + a(start + 1) + ... + a(stop)
  answer = 0
  for k in range(start, stop+1):
    answer += a(k)
  return answer

def f(n):
  return 1/math.factorial(n)

def g(n):
  return (-1)**n/(2*n+1)

print(sum_series(0, 20, f))
print(sum_series(0, 10, lambda x:1/math.factorial(x)))

print(4* sum_series(0, 1000, g))
```

Write a function to test if n is prime, and then call it from a for loop.
Talk about parameter passing, where and how to call the function

```python
def is_prime(n):
  #Returns true only when n is a prime number
  answer = True
  if n > 2 and n % 2 == 0:
    answer = False  # n is even
  else:
    factor = 3
    while answer and factor * factor <= n:
      if n % factor == 0:
        answer = False
      else:
        factor += 2
  return answer

# Create a list of prime numbers less than 1000 using a for loop
primes = []
for k in range(2, 1000):
  if is_prime(k):
    primes.append(k)

# Create a list of prime numbers less than 1000 using list comprehension
primes2 = [k for k in range(2, 1000) if is_prime(k)]

#Display the lists in various ways
print(primes2)
print(f'----------')
print(f'The prime numbers less than 1000 are: ')
for k in range(len(primes)):
  print(primes[k], end = ' ')
  if k%10 == 0 and k > 0:
    print()
print()
print(f'----------')

print(f'The distribution of prime numbers less than 1000: ')
index = 0
for row in range(10):
  while index < len(primes) and 100*row < primes[index] < 100*(row+1):
    print(f'{primes[index]:3} ', end = '')
    index += 1
  print()
```

Write a solution to the reverse complement DNA problem using functions

```python
import random

def complement(letter):
  if letter == 'A':
    return 'T'
  elif letter == 'T':
    return 'A'
  elif letter == 'C':
    return 'G'
  elif letter == 'G':
    return 'C'
  else:
    return 'X'

def reverse_complement(astring):
  answer = ''
  for symbol in astring:
    answer = complement(symbol) +answer
    #print(f'astring:  {astring},   answer: {answer}')  #for debugginh
  return answer

ex1 = 'acgt'.upper()
print(f'{ex1} has reverse complement {reverse_complement(ex1)}')

ex2 = ''.join([random.choice('ACTG') for k in range(10)])
print(f'{ex2} has reverse complement {reverse_complement(ex2)}')
```

```python
import random

def g_concentration(astring):
  random.seed(1)
  length = len(astring)
  g_count = 0
  for letter in astring:
    if letter == 'G':
      g_count+=1
  return g_count / length

m = 50
random_dna = ''.join([random.choice('ACTG') for k in range(m)])
print(f'The g-concentration of {random_dna} is {g_concentration(random_dna)}.')
```
