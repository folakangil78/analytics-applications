Numerical Integration:  Functions as parameters to other functions

```python
import numpy as np

def numerical_integral(a, b, f, n):
  dx = (b-a)/n
  answer = 0
  x = a
  for k in range(n):
    answer += f(x)*dx
    x += dx
  return answer

def f1(x):
  return x*x

def f2(x):
  return np.exp(-x*x)


num_rect = 10000
print(f'The integral of x^2 from x = {0} to x = {3} using {num_rect} rectangles is {numerical_integral(0, 3, f1, num_rect):.4f}')

print(f'The integral of e^(-x^2) from x = {0} to x = {3} using {num_rect} rectangles is {numerical_integral(0, 3, f2, num_rect):.4f}')

print(f'The integral of sin(x) from x = {0} to x = {np.pi:.4f} using {num_rect} rectangles is {numerical_integral(0, np.pi, np.sin, num_rect):.4f}')
```

Write a function to test if n is prime, and then call it from a for loop.
Talk about parameter passing, where and how to call the function

```python
def is_prime(n):
  #Returns true only when n is a prime number
  answer = True
  if n > 2 and n % 2 == 0:
    answer = False  # n is even
  else:
    factor = 3
    while answer and factor * factor <= n:
      if n % factor == 0:
        answer = False
      else:
        factor += 2
  return answer

# Create a list of prime numbers less than 1000 using a for loop
primes = []
for k in range(2, 1000):
  if is_prime(k):
    primes.append(k)

# Create a list of prime numbers less than 1000 using list comprehension
primes2 = [k for k in range(2, 1000) if is_prime(k)]

#Display the lists in various ways
print(primes2)
print(f'----------')
print(f'The prime numbers less than 1000 are: ')
for k in range(len(primes)):
  print(primes[k], end = ' ')
  if k%10 == 0 and k > 0:
    print()
print()
print(f'----------')

print(f'The distribution of prime numbers less than 1000: ')
index = 0
for row in range(10):
  while index < len(primes) and 100*row < primes[index] < 100*(row+1):
    print(f'{primes[index]:3} ', end = '')
    index += 1
  print()
```

Write a solution to the reverse complement DNA problem using functions

```python
import random

def complement(letter):
  if letter == 'A':
    return 'T'
  elif letter == 'T':
    return 'A'
  elif letter == 'C':
    return 'G'
  elif letter == 'G':
    return 'C'
  else:
    return 'X'

def reverse_complement(astring):
  answer = ''
  for symbol in astring:
    answer = complement(symbol) +answer
    #print(f'astring:  {astring},   answer: {answer}')  #for debugginh
  return answer

ex1 = 'acgt'.upper()
print(f'{ex1} has reverse complement {reverse_complement(ex1)}')

ex2 = ''.join([random.choice('ACTG') for k in range(10)])
print(f'{ex2} has reverse complement {reverse_complement(ex2)}')
```

A menu driven loop to perform tasks (on DNA strings).  Notice function stubs

```python
def show_menu():
  print('Welcome')
  print('Generate a random dna stiring')
  print('Compute the g-concentration')
  print('task 3')
  # show a menu of options

def get_choice():
  # return the user's choice of options
  answer = input('What do you choose?')
  return answer

def random_dna(n):
  return ''.join([random.choice('ACTG') for k in range(n)])

def perform_action(a):
  if a == 'random':
    return random_dna(10)

def main():
  show_menu()
  user_choice = get_choice()
  while user_choice[0] not in ['q', 'Q']:
    perform_action(user_choice)
    show_menu()
    user_choice = get_choice()
  print('Goodbye')

main()
```
