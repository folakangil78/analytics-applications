```python
import random

def game(stake, goal):
  # Input parameters:  stake, the player's initial amount of money in dollars
  #                    goal, the dollar amount that the player wants to reach - game ends if goal is reached or surpassed
  #                    win_prob a float number between 0 and 1 given the probability of getting heads in the coin flip
  # Return values:  (n, x) where n is the number of times the coin was flipped and
  #                    x = 1 if the player reaches or exceeds their goal, x = 0 if the game ends with the player losing all their money and not reaching their goal
  # The game:  Repeatedly flip the coin.  If it lands on heads, the player recieves $1,  If it lands on tails, the plays loses $1
  # Play continues until the play runs out of money, or reaches or exceeds their goal.

  num_of_bets = 0

  if stake == 0:
    return num_of_bets, 0

  while stake > 0 and stake < goal:
    choice = random.choice(["Heads", "Tails"])
    if choice == "Heads":
      stake+=1
    else:
      stake-=1
    num_of_bets+=1

  if stake == 0:
    return num_of_bets, 0
  elif stake >= goal:
    return num_of_bets, 1 # returning number of bets made and returning 1 if player wins
                          # or 0 if he goes broke


def simulate(num_trials, stake, goal):
  # Input parameters:  num_trials, a positive int value for the number of times to repeat game
  #                    stake, the player's initial amount of money in dollars
  #                    goal, the dollar amount that the player wants to reach - game ends if goal is reached or surpassed
  #                    win_prob a float number between 0 and 1 given the probability of getting heads in the coin flip.  The default value is 0.5 for a fair coin
  # The simulation:  Repeat gameA num_trials many times and determine how likely it is that the player will win
  # Return values: (pw, enf)  where pw is the estimated probability that the player will win  = num_wins/num_trials
  #                    and enf is the estimated number of flips untils the game ends

  num_wins = 0
  num_of_bets = 0
  for i in range(num_trials):
    total_bets, win_or_loss = game(stake, goal)
    num_of_bets = num_of_bets + total_bets
    if win_or_loss == 1:
      num_wins = num_wins + 1

  pw = num_wins / num_trials # estimated prob that gambler wins
  enf = num_of_bets / num_trials # estimated number of flips until game ends whether he wins/loses
  #no specified requ. to round avg # of flips

  return pw, enf



def main():
  print("Welcome to the Gambler's Ruin Program")
  s = float(input("What is the player's original stake: "))
  g = float(input("What is the player's goal: "))
  n = int(input('How many times should we run the experiment? '))

  prob_w, num_games = simulate(n, s, g)
  print(f'Using {n:,} trials,  stake = {s}, and goal = {g} ')
  print(f'the probability of winning is {prob_w} and the average number of flips to end the game is {num_games}')

main()
```

```python
#Problem 2
import numpy as np

def numerical_integral(a, b, f, n):
  # This is the Left-hand Riemann Sum - do not change this
  dx = (b-a)/n
  answer = 0
  x = a
  for k in range(n):
    answer += f(x)*dx
    x += dx
  return answer

def trapezoid_integral(a, b, f, n):
  trapezoid_const = 0.5 # avoids magic vals
  prelim_trapezoid_area = trapezoid_const * (f(a) + f(b)) # taken from trap formula wiki
  # deals with fxn values for start and end bounds, allows for loop to start at 1

  bound_diff = (b - a) / n # width of interval

  for j in range(1, n): # sums result (height) of func at each trapezoid
    x = a + j * bound_diff # calculates x val relative to beginning 'a' bound and width
    prelim_trapezoid_area += f(x)

  return prelim_trapezoid_area * bound_diff # multiply summed heights from loop by (consistent) width

def f1(x):
  return x*x

def f2(x):
  return np.exp(-x*x)

def main():
  num_rect = 1000
  print(f'The integral of x^2 from x = {0} to x = {3} using {num_rect} rectangles is {numerical_integral(0, 3, f1, num_rect):.4f}')
  print(f'The integral of x^2 from x = {0} to x = {3} using {num_rect} trapezoids is {trapezoid_integral(0, 3, f1, num_rect):.4f}')
  print()
  print(f'The integral of e^(-x^2) from x = {0} to x = {3} using {num_rect} rectangles is {numerical_integral(0, 3, f2, num_rect):.4f}')
  print(f'The integral of e^(-x^2) from x = {0} to x = {3} using {num_rect} trapezoids is {trapezoid_integral(0, 3, f2, num_rect):.4f}')
  print()
  print(f'The integral of sin(x) from x = {0} to x = {np.pi:.4f} using {num_rect} rectangles is {numerical_integral(0, np.pi, np.sin, num_rect):.4f}')
  print(f'The integral of sin(x) from x = {0} to x = {np.pi:.4f} using {num_rect} trapezoids is {trapezoid_integral(0, np.pi, np.sin, num_rect):.4f}')

main()
```
